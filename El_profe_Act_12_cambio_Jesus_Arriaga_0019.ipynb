## Las variables
1. CP --> Costo de producto
2. CD --> Cantidad de dinero
3. Cambio

## Formulas
$Cambio = CD - CP$

## Diagrama de flujo LicidChart
<img src="https://raw.githubusercontent.com/JesusArriaga128/Calcular-cambio-if-else-/refs/heads/main/Df%20if%20else%20Cambio%202.png" width="350" height="500">



```python
# Para cuando es verdadero (Si)
# Calculando cambio usando if else
# Datos de entrada CD , CP
# Mostrar Cambio si es falso
# si es verdadero mostrar No alcanza
CP = float(input("Ingrese el costo del producto: "))
CD = float(input("Ingrese la cantidad de dinero: "))
if CP > CD:
    print("No alcanza")
else:
    Cambio = CD - CP
    print("El cambio es: ", Cambio)
print("fin de programa")
print("Programa realizado Jesus Arriaga NL = 5")
```

```python
# Para cuando es falso (No)
# Calculando cambio usando if else
# Datos de entrada CD , CP
# Mostrar Cambio si es falso
# si es verdadero mostrar No alcanza
CP = float(input("Ingrese el costo del producto: "))
CD = float(input("Ingrese la cantidad de dinero: "))
if CP > CD:
    print("No alcanza")
else:
    Cambio = CD - CP
    print("El cambio es: ", Cambio)
print("fin de programa")
print("Programa realizado Jesus Arriaga NL = 5")
```
